```python
# Variables para los rect치ngulos m치s cercanos en las esquinas superior derecha e izquierda
top_right_rectangle = None
bottom_right_rectangle = None
min_distance_top_right = float('inf')
min_distance_bottom_right = float('inf')

# Lista para almacenar las coordenadas de los rect치ngulos
coordinates_ranges = []

# Iterar a trav칠s de los contornos y seleccionar los rect치ngulos m치s cercanos
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    center_x = x + w // 2
    center_y = y + h // 2

    # Calcular la distancia desde el centro del rect치ngulo de referencia
    distance = np.sqrt((center_x - (x_ref + x2_ref / 2)) ** 2 + (center_y - (y_ref + y2_ref / 2)) ** 2)

    # Seleccionar el rect치ngulo m치s cercano a la esquina superior derecha
    if center_x > (x2_ref) and center_y < (y2_ref) and distance < min_distance_top_right:
        top_right_rectangle = (x, y, w, h)
        min_distance_top_right = distance

    # Seleccionar el rect치ngulo m치s cercano a la esquina inferior derecha
    if center_x > (x2_ref) and center_y > (y2_ref) and distance < min_distance_bottom_right:
        bottom_right_rectangle = (x, y, w, h)
        min_distance_bottom_right = distance

# Agregar las coordenadas de los rect치ngulos a la lista
if top_right_rectangle is not None:
    coordinates_ranges.append(top_right_rectangle)
if bottom_right_rectangle is not None:
    coordinates_ranges.append(bottom_right_rectangle)

# Crear una copia de la imagen RGB original para dibujar los rect치ngulos
image_rgb_copy = image_rgb.copy()

# Dibujar el rect치ngulo de referencia en blanco
cv2.rectangle(image_rgb_copy, (x_ref, y_ref), (x2_ref, y2_ref), (255, 255, 255), 2)  # Blanco: BGR (255, 255, 255)

# Dibujar los rect치ngulos m치s cercanos en verde
for rect in coordinates_ranges:
    x, y, w, h = rect
    cv2.rectangle(image_rgb_copy, (x, y), (x + w, y + h), (0, 255, 0), 2)  # Verde: BGR (0, 255, 0)

# Mostrar la imagen con los rect치ngulos resaltados
plt.show()
plt.imshow(image_rgb_copy)
plt.title('Rect치ngulo de Referencia en Blanco, Rect치ngulos M치s Cercanos en Verde')
plt.axis('off')  # Ocultar los ejes
plt.show()

#print(x_ref,y_ref)
#print(x2_ref,y2_ref)

# Imprimir las coordenadas de los rect치ngulos
print("Coordenadas de los rect치ngulos:", coordinates_ranges)
```

### F칩rmula de Interpolaci칩n Lineal

La interpolaci칩n lineal se puede utilizar para asignar temperaturas a cada valor de p칤xel en una barra de colores. Aqu칤 est치 la f칩rmula:

$ \text{Temperatura} = \text{TemperaturaMinima} + \left(\frac{\text{ValorDelPixel} - \text{ValorMinimo}}{\text{ValorMaximo} - \text{ValorMinimo})}\right) \times (\text{TemperaturaMaxima} - \text{TemperaturaMinima}) $

Donde:
- "Temperatura" es la temperatura asociada al valor del p칤xel en la barra.
- "Temperatura m칤nima" es el valor m칤nimo de temperatura que tiene la termograf칤a.
- "Valor del p칤xel" es el valor de intensidad del p칤xel en la escala de grises de la barra.
- "Valor m칤nimo" es el valor m칤nimo de intensidad de p칤xeles en la barra (puede ser 0 si es escala de grises).
- "Rango de valores de p칤xeles" es la diferencia entre el valor m치ximo y el valor m칤nimo de intensidad de p칤xeles en la barra.
- "Temperatura m치xima" es el valor m치ximo de temperatura que tiene la termograf칤a.

Esta f칩rmula permite asignar temperaturas a cada p칤xel en la barra de colores bas치ndote en su intensidad de escala de grises.


```python
temperaturas_promedio_filtradas=[]
# Calcular el rango de temperaturas
min_temp = bottom_temperature_expanded
max_temp = top_temperature_expanded

# Obtener las coordenadas de la columna vertical
column_x = x_ref  # Suponemos que la columna est치 en la coordenada x_ref
column_y_start = y_ref
column_y_end = y2_ref
data_dictRGB={}

# Cargar la imagen termogr치fica
image = cv2.imread('1.jpg')

# Convertir la imagen a espacio de color RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Obtener la columna de un p칤xel de grosor desde la parte superior hasta la inferior
for y in range(column_y_start, column_y_end + 1):
    # Obtener el valor del componente rojo en el p칤xel actual
    r_value = image_rgb[y, column_x, 0]  # Componente R en el canal 0 (rojo)
    
    # Calcular la temperatura asociada al p칤xel en funci칩n de su posici칩n vertical
    # Suponemos que el cambio de temperatura es lineal de abajo hacia arriba
    temp = min_temp + (max_temp - min_temp) * (y - column_y_start) / (column_y_end - column_y_start)
    
    # Almacenar los valores en el diccionario
    data_dictRGB[y] = {'R_mask': r_value, 'Temperature': temp}
    temperaturas_promedio_filtradas.append(temp)

# Imprimir el diccionario resultante
print(data_dictRGB)
```
