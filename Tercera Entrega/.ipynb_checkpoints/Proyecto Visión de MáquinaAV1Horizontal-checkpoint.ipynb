```python
# Aplicar umbralizaci칩n a la imagen en escala de grises
ret, thresh1 = cv2.threshold(gray, 250, 255, cv2.THRESH_BINARY)

# Mostrar la imagen umbralizada
plt.show()
plt.imshow(thresh1)
plt.title('Imagen Umbralizada')
plt.axis('off')  # Ocultar los ejes
plt.show()

```

```python
# Imprimir la matriz de umbralizaci칩n
print(thresh1)

# Crear un kernel para operaciones morfol칩gicas
kernel = np.ones((20, 20), np.uint8)

# Realizar una operaci칩n de cierre morfol칩gico en la imagen umbralizada
joined = cv2.morphologyEx(thresh1, cv2.MORPH_CLOSE, kernel)

# Encontrar contornos en la imagen despu칠s del cierre morfol칩gico
contours, _ = cv2.findContours(joined, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Mostrar la imagen despu칠s del cierre morfol칩gico
plt.show()
plt.imshow(joined)
plt.title('Imagen despu칠s del Cierre Morfol칩gico')
plt.axis('off')  # Ocultar los ejes
plt.show()

```

```python
# Variables para los rect치ngulos m치s cercanos en las esquinas superior derecha e izquierda
top_right_rectangle = None
bottom_right_rectangle = None
min_distance_top_right = float('inf')
min_distance_bottom_right = float('inf')

# Lista para almacenar las coordenadas de los rect치ngulos
coordinates_ranges = []

# Iterar a trav칠s de los contornos y seleccionar los rect치ngulos m치s cercanos
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    center_x = x + w // 2
    center_y = y + h // 2

    # Calcular la distancia desde el centro del rect치ngulo de referencia
    distance = np.sqrt((center_x - (x_ref + x2_ref / 2)) ** 2 + (center_y - (y_ref + y2_ref / 2)) ** 2)

    # Seleccionar el rect치ngulo m치s cercano a la esquina superior derecha
    if center_x > (x2_ref) and center_y < (y2_ref) and distance < min_distance_top_right:
        top_right_rectangle = (x, y, w, h)
        min_distance_top_right = distance

    # Seleccionar el rect치ngulo m치s cercano a la esquina inferior derecha
    if center_x > (x2_ref) and center_y > (y2_ref) and distance < min_distance_bottom_right:
        bottom_right_rectangle = (x, y, w, h)
        min_distance_bottom_right = distance

# Agregar las coordenadas de los rect치ngulos a la lista
if top_right_rectangle is not None:
    coordinates_ranges.append(top_right_rectangle)
if bottom_right_rectangle is not None:
    coordinates_ranges.append(bottom_right_rectangle)

# Crear una copia de la imagen RGB original para dibujar los rect치ngulos
image_rgb_copy = image_rgb.copy()

# Dibujar el rect치ngulo de referencia en blanco
cv2.rectangle(image_rgb_copy, (x_ref, y_ref), (x2_ref, y2_ref), (255, 255, 255), 2)  # Blanco: BGR (255, 255, 255)

# Dibujar los rect치ngulos m치s cercanos en verde
for rect in coordinates_ranges:
    x, y, w, h = rect
    cv2.rectangle(image_rgb_copy, (x, y), (x + w, y + h), (0, 255, 0), 2)  # Verde: BGR (0, 255, 0)

# Mostrar la imagen con los rect치ngulos resaltados
plt.show()
plt.imshow(image_rgb_copy)
plt.title('Rect치ngulo de Referencia en Blanco, Rect치ngulos M치s Cercanos en Verde')
plt.axis('off')  # Ocultar los ejes
plt.show()

#print(x_ref,y_ref)
#print(x2_ref,y2_ref)

# Imprimir las coordenadas de los rect치ngulos
print("Coordenadas de los rect치ngulos:", coordinates_ranges)
```

### F칩rmula de Interpolaci칩n Lineal

La interpolaci칩n lineal se puede utilizar para asignar temperaturas a cada valor de p칤xel en una barra de colores. Aqu칤 est치 la f칩rmula:

$ \text{Temperatura} = \text{TemperaturaMinima} + \left(\frac{\text{ValorDelPixel} - \text{ValorMinimo}}{\text{ValorMaximo} - \text{ValorMinimo})}\right) \times (\text{TemperaturaMaxima} - \text{TemperaturaMinima}) $

Donde:
- "Temperatura" es la temperatura asociada al valor del p칤xel en la barra.
- "Temperatura m칤nima" es el valor m칤nimo de temperatura que tiene la termograf칤a.
- "Valor del p칤xel" es el valor de intensidad del p칤xel en la escala de grises de la barra.
- "Valor m칤nimo" es el valor m칤nimo de intensidad de p칤xeles en la barra (puede ser 0 si es escala de grises).
- "Rango de valores de p칤xeles" es la diferencia entre el valor m치ximo y el valor m칤nimo de intensidad de p칤xeles en la barra.
- "Temperatura m치xima" es el valor m치ximo de temperatura que tiene la termograf칤a.

Esta f칩rmula permite asignar temperaturas a cada p칤xel en la barra de colores bas치ndote en su intensidad de escala de grises.


```python
# Obtener las coordenadas de la columna vertical
column_x = x_ref  # Suponemos que la columna est치 en la coordenada x_ref
column_y_start = y_ref
column_y_end = y2_ref

# Cargar la imagen termogr치fica
image = cv2.imread('1.jpg')

# Convertir la imagen a espacio de color HSV
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Calcular el rango de temperaturas
min_temp = bottom_temperature_expanded
max_temp = top_temperature_expanded

# Crear un diccionario para almacenar los valores de la m치scara H y las temperaturas asociadas de la barra
data_dict = {}

# Obtener la columna de un p칤xel de grosor desde la parte superior hasta la inferior
for y in range(column_y_start, column_y_end + 1):
    # Obtener el valor de la m치scara H en el p칤xel actual
    h_value = image_hsv[y, column_x, 0]
    
    # Calcular la temperatura asociada al p칤xel en funci칩n de su posici칩n vertical
    # Suponemos que el cambio de temperatura es lineal de abajo hacia arriba
    temp = min_temp + (max_temp - min_temp) * (y - column_y_start) / (column_y_end - column_y_start)
    
    # Almacenar los valores en el diccionario
    data_dict[y] = {'H_mask': h_value, 'Temperature': temp}
    

# Imprimir el diccionario resultante
print(data_dict)
```

```python
import cv2
import numpy as np

# Cargar la imagen termogr치fica
image = cv2.imread('1.jpg')

# Cargar las temperaturas en cada p칤xel de la imagen original
temperature_data = data_dict  # Suponemos que data_dict contiene las temperaturas

# Crear una nueva matriz para la imagen de matiz con las mismas dimensiones que la imagen original
hue_image = np.zeros_like(image, dtype=np.uint8)

# Recorrer la imagen original y asignar los valores de matiz basados en las temperaturas
for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        hue_value = image[y, x, 0]  # Valor de matiz en el p칤xel actual

        # Buscar la temperatura correspondiente en data_dict
        temperature = None
        for temp_y, temp_data in temperature_data.items():
            if hue_value >= temp_data['H_mask']:
                temperature = int(temp_data['Temperature'])
            else:
                break

        if temperature is not None:
            hue_image[y, x, 0] = temperature  # Asignar la temperatura como matiz en la nueva imagen

            # Muestra la imagen original junto a la m치scara
plt.figure(figsize=(12, 6))
plt.imshow(hue_image)  # Usamos 'jet' para el mapa de colores
plt.title('Mapa de Matiz')
plt.axis('off')
plt.show()

    
```

```python
# Cargar la matriz de valores de matiz (mismas dimensiones que la imagen original)
hue_matrix = hue_image

# Escalar los valores de matiz para que est칠n en el rango [0, 255]
scaled_hue = np.interp(hue_matrix, (hue_matrix.min(), hue_matrix.max()), (0, 255)).astype(np.uint8)

# Crear una matriz vac칤a para la imagen en espacio HSV
image_hsv = np.zeros((scaled_hue.shape[0], scaled_hue.shape[1], 3), dtype=np.uint8)

# Asignar los valores de matiz a la matriz HSV
image_hsv[:, :, 0] = scaled_hue[:, :, 0]  # Asignar el canal de matiz de scaled_hue a image_hsv
image_hsv[:, :, 1] = 255  # Establecer la saturaci칩n a 255 (m치xima saturaci칩n)
image_hsv[:, :, 2] = 255  # Establecer el valor (brillo) a 255 (m치ximo brillo)


# Convertir la imagen de espacio HSV a espacio BGR para visualizarla
image_bgr = cv2.cvtColor(image_hsv, cv2.COLOR_HSV2BGR)

# Guardar la imagen resultante en un archivo
cv2.imwrite('imagen_hsv_resultante.png', image_bgr)

# Plotear la imagen
plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB))  # Convertir a formato RGB antes de mostrar
plt.axis('off')
plt.title('Imagen en Espacio HSV con Matiz Modificado')
plt.show()
```

```python
# Lee la imagen termogr치fica
image = cv2.imread('1.jpg')

areas_filtradas = []
perimetros_filtrados = []
centroides_filtrados = []
temperaturas_promedio_filtradas = []
id_filtrados=[]


# Itera sobre los rect치ngulos y establece a 0 los p칤xeles dentro de cada uno
for rect in rectangles_coordinates:
    (x, y), (x_end, y_end) = rect
    image[y:y_end, x:x_end, 2] = 0

# Extrae el canal rojo
red_channel = image[:, :, 2]

# Aplica el m칠todo de Otsu para encontrar el umbral autom치ticamente
_, otsu_threshold = cv2.threshold(red_channel, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Encuentra los contornos en la imagen umbralizada
contours, _ = cv2.findContours(otsu_threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Obtener la columna de un p칤xel de grosor desde la parte superior hasta la inferior
for y in range(y_ref, y2_ref + 1):
    # Obtener el valor del componente rojo en el p칤xel actual
    r_value = image_rgb[y, x_ref, 0]  # Componente R en el canal 0 (rojo)
    
    # Calcular la temperatura asociada al p칤xel en funci칩n de su posici칩n vertical
    # Suponemos que el cambio de temperatura es lineal de abajo hacia arriba
    temp = min_temp + (max_temp - min_temp) * (y - y_ref) / (y2_ref - y_ref)
    
    # Almacenar los valores en el diccionario
    data_dictRGB[y] = {'R_mask': r_value, 'Temperature': temp}

# Itera sobre los contornos y dibuja rect치ngulos rojos
for i, contour in enumerate(contours, start=1):
    # Calcula el rect치ngulo delimitador
    x, y, w, h = cv2.boundingRect(contour)

    # Calcula el 치rea
    area = cv2.contourArea(contour)
    

    # Calcula el per칤metro
    perimeter = cv2.arcLength(contour, True)
    


    # Calcula el centroide
    M = cv2.moments(contour)
    if M["m00"] != 0:
        centroid_x = int(M["m10"] / M["m00"])
        centroid_y = int(M["m01"] / M["m00"])
        centroid = (centroid_x, centroid_y)

        
        
    else:
        centroid = None

    # Descarta las 치reas con centroide None y 치rea menor a 100
    if centroid is not None and area >= 100 and centroid_y in data_dictRGB:
        # Dibuja el rect치ngulo en la imagen original (rojo)
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 2)

        # Muestra informaci칩n en la consola
        print(f"Regi칩n {i} - 츼rea: {area}, Per칤metro: {perimeter}, Centroid: {centroid}")
        areas_filtradas.append(area)
        perimetros_filtrados.append(perimeter)
        centroides_filtrados.append(centroid) 
        id_filtrados.append(i)
        

        import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Crear una figura y un objeto de ejes
fig, ax = plt.subplots(figsize=(10, 10))

# Mostrar la imagen con las regiones filtradas
ax.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

# Iterar sobre las regiones filtradas y dibujar rect치ngulos con informaci칩n
for i in range(len(areas_filtradas)):
    area = areas_filtradas[i]
    perimeter = perimetros_filtrados[i]
    centroid = centroides_filtrados[i]

    
    # Dibujar el rect치ngulo
    rect = patches.Rectangle((centroid[0] - w/2, centroid[1] - h/2), w, h, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)

    # Mostrar informaci칩n de 치rea, per칤metro y centroide
    plt.text(centroid[0] - w/2, centroid[1] - h/2 - 10, f"Area: {area:.2f}\nPerimeter: {perimeter:.2f}\nCentroid: {centroid}", color='r', fontsize=8, bbox=dict(facecolor='white', alpha=0.7))

# Ajustar la escala de los ejes
ax.autoscale()

# Mostrar la imagen
plt.title('Regiones Filtradas (츼rea >= 100)')
plt.axis('off')
plt.show()

```

```python
# Cargar la imagen termogr치fica
image = cv2.imread('1.jpg')

# Convertir la imagen a espacio de color RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Lista para almacenar las tuplas de ancho y altura
rectangles_sizes = []

# Crear una imagen en blanco para visualizar las regiones
regions_image = np.zeros_like(image_rgb)

# Obtener la columna de un p칤xel de grosor desde la parte superior hasta la inferior
for y in range(y_ref, y2_ref + 1):
    # Obtener el valor del componente rojo en el p칤xel actual
    r_value = image_rgb[y, x_ref, 0]  # Componente R en el canal 0 (rojo)
    
    # Calcular la temperatura asociada al p칤xel en funci칩n de su posici칩n vertical
    # Suponemos que el cambio de temperatura es lineal de abajo hacia arriba
    temp = min_temp + (max_temp - min_temp) * (y - y_ref) / (y2_ref - y_ref)
    
    # Almacenar los valores en el diccionario
    data_dictRGB[y] = {'R_mask': r_value, 'Temperature': temp}

# Iterar sobre las regiones y asignar temperatura promedio
for i, contour in enumerate(contours, start=1):
    # Calcula el rect치ngulo delimitador
    x, y, w, h = cv2.boundingRect(contour)

    # Calcula el 치rea
    area = cv2.contourArea(contour)

    # Si el 치rea es mayor a 100, realiza el procesamiento
    if area > 100:
        # Calcula el centroide
        M = cv2.moments(contour)
        if M["m00"] != 0:
            centroid_x = int(M["m10"] / M["m00"])
            centroid_y = int(M["m01"] / M["m00"])
            centroid = (centroid_x, centroid_y)
        else:
            centroid = None

        # Asigna temperatura promedio a la regi칩n
        if centroid is not None and centroid_y in data_dictRGB:
            # Obtiene la temperatura del diccionario
            temperature = data_dictRGB[centroid_y]['Temperature']
            

            # Asigna la temperatura promedio a la regi칩n
            average_temperature = np.mean([temperature for _ in range(int(centroid_y - h / 2), int(centroid_y + h / 2) + 1)])

            # Muestra informaci칩n en la consola
            print(f"Regi칩n {i} - Temperatura promedio: {average_temperature}")
            temperaturas_promedio_filtradas.append(average_temperature)
            # Almacenar las tuplas de ancho y altura en la lista
            rectangles_sizes.append((w, h))


            # Dibuja el rect치ngulo en la imagen de las regiones
            cv2.rectangle(regions_image, (x, y), (x + w, y + h), (0, 0, 255), 2)
            
            

            

# Configuraci칩n de la visualizaci칩n de las regiones filtradas
fig, ax = plt.subplots(figsize=(12, 10))

# Imagen original detr치s
plt.imshow(image_rgb)
plt.title('Regiones Filtradas (츼rea >= 100) con Temperatura')
plt.axis('off')

# Iterar sobre las regiones filtradas y dibujar rect치ngulos con informaci칩n
for i in range(len(areas_filtradas)):
    area = areas_filtradas[i]
    perimeter = perimetros_filtrados[i]
    centroid = centroides_filtrados[i]
    temperature = temperaturas_promedio_filtradas[i]

    # Dibujar el rect치ngulo
    rect = patches.Rectangle((centroid[0] - w/2, centroid[1] - h/2), w, h, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)

    # Mostrar informaci칩n de 치rea, per칤metro y centroide
    plt.text(centroid[0] - w/2, centroid[1] - h/2 - 10, f"Area: {area:.2f}\nPerimeter: {perimeter:.2f}\nCentroid: {centroid}\nTemperature: {temperature:.2f}", color='r', fontsize=8, bbox=dict(facecolor='white', alpha=0.7))

# Ajustar la escala de los ejes
ax.autoscale()

# Mostrar la imagen con rect치ngulos y la informaci칩n adicional
plt.show()

# Mostrar ambas im치genes en una sola figura
plt.figure(figsize=(12, 10))

# Imagen de las regiones
plt.subplot(221), plt.imshow(regions_image)
plt.title('Regiones con Temperatura Promedio (츼rea > 100)')
plt.axis('off')

# Imagen umbralizada con Otsu
plt.subplot(222), plt.imshow(cv2.cvtColor(red_image_thresholded_otsu, cv2.COLOR_BGR2RGB))
plt.title('Imagen Umbralizada (Otsu)')
plt.axis('off')

plt.show()
```
